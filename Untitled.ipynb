```python
import requests as req
import bs4

def get_wiki(url):
    with req.get(url) as r:
        soup = bs4.BeautifulSoup(r.content, 'html.parser')

    main_content = soup.find("main")

    # Classes and IDs to exclude (blacklist)
    exclude_classes = ['noexcerpt', 'noprint', 'dablink']
    exclude_ids = ['mw-panel-toc','p-lang-btn-checkbox']  # Add any specific IDs you want to exclude

    # Text tags to include
    text_tags = ['p', 'h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'li', 'span']

    # Set to keep track of excluded elements
    excluded_elements = set()

    def add_to_excluded(element):
        # Recursively add element and all its descendants to the excluded set
        if element not in excluded_elements:
            excluded_elements.add(element)
            for child in element.descendants:
                excluded_elements.add(child)

    # Add excluded elements and their descendants to the set
    for cls in exclude_classes:
        for elem in main_content.find_all(class_=cls):
            add_to_excluded(elem)
    for id_val in exclude_ids:
        elem = main_content.find(id=id_val)
        if elem:
            add_to_excluded(elem)

    # Extract text from allowed elements
    extracted_texts = []
    for element in main_content.find_all(lambda x: x not in excluded_elements):
        if element not in excluded_elements:
            extracted_texts.append(element.text)
            add_to_excluded(element)  # Add the element and its descendants to the set

    return "\n\n".join(extracted_texts)

```
